<h4>Подключение библиотек</h4>

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

<h4>Написание метода, создающий массив, содержащий выпавшие грани кубика за n раз</h4>

```python
def cube(n: int) -> list:
    data = []
    while len(data) < n:
        data.append(random.randint(1,6))
    return data

list1 = cube(100)
list2 = cube(1000)
list3 = cube(10000)
list4 = cube(1000000)
```

<h4>Вывод гистограмм соотношения</h4>

```python
plt.hist(list1, color='b', bins=np.arange(1, 8) - 0.5, rwidth=0.8, alpha=0.4)
plt.title('Соотношение выпадений граней кубика за 100 бросков')
plt.show()

plt.hist(list2, color='r',  bins=np.arange(1, 8) - 0.5, rwidth=0.8, alpha=0.4)
plt.title('Соотношение выпадений граней кубика за 1000 бросков')
plt.show()

plt.hist(list3, color='g',  bins=np.arange(1, 8) - 0.5, rwidth=0.8, alpha=0.4)
plt.title('Соотношение выпадений граней кубика за 10 000 бросков')
plt.show()

plt.hist(list4, color='purple',  bins=np.arange(1, 8) - 0.5, rwidth=0.8, alpha=0.4)
plt.title('Соотношение выпадений граней кубика за 1 000 000 бросков')
plt.show()
```

<h3>Вывод</h3>
Из графиков видно, что с увеличением количества бросков вероятность выпадения каждого из чисел уравниваются.
Опыт доказывает закон больших чисел.
